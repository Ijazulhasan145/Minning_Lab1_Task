## Task2


```python
monthly_sales = pd.Series([200, 300, None, 150, 400, None, 350, 300, 200, None, 450,500])
monthly_sales
```

```python
# Calculate AVG
avg_sale=monthly_sales.mean()
avg_sale
```

```python
# Fill missing Value with AVG sale
monthly_sale_fill = monthly_sales.fillna(avg_sale)
monthly_sale_fill
```

```python
# Fill missing Value with two more months to the end Nan
monthly_sales_reindexed = monthly_sale_fill.reindex(range(14))
monthly_sales_reindexed
```

```python
# Fill missing Value with 0
monthly_sales_zero = monthly_sales_reindexed.fillna(0)
monthly_sales_zero
```

```python
sensor_data = pd.DataFrame({ 'Sensor_1': [25, 145, np.nan, 152, 130, np.nan, 40, 60, 180,
90], 'Sensor_2': [np.nan, 60, 75, 140, np.nan, 160, 155, 80, 110, 145], 'Sensor_3': [100,
np.nan, 135, 165, 45, 150, np.nan, 155, 120, 170], 'Sensor_4': [120, 135, 140, np.nan, 125,
180, np.nan, 70, 160, 150], 'Sensor_5': [140, np.nan, 155, 60, 175, 130, 95, np.nan, 150,
115] })
sensor_data
```

```python
# Fill missing value
sensor_data_filled = sensor_data.apply(lambda x: x.fillna(x.mean()), axis=0)
sensor_data_filled
```

```python
# control threshold
sensor_data_threshold = sensor_data_filled.applymap(lambda x: min(x, 150))
sensor_data_threshold
```

```python
pixel_values = np.array([23, 45, 167, 89, 34, 120, 56, 200, 78, 144, 34, 255, 189, 67, 90,
160, 128, 45, 210, 75, 44, 88, 77, 150, 99])
pixel_values
```

```python
# step1:Reshape 5x5
image_matrix = pixel_values.reshape(5,5)
image_matrix
```

```python
# Step 2: Cap values above 150( pixel values above 150 needs to be set to 150)
image_matrix = np.clip(image_matrix, 0, 150)  
image_matrix
```

```python
# Step 3: Extract the central 3x3 region
central_region = image_matrix[1:4, 1:4]
central_region
```
